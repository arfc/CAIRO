# Processes the Accumulated Solar Data for Several Places

The data was downloaded from the NREL National Solar Radiation Database (NSRD)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import glob
import os
```

```python
# the list of locations

wind_locations = ['san_bernadino',
                  'santa_fe',
                  'dallas', 
                  'lincoln',
                  'mansfield',
                  'syracuse']

solar_locations = ['san_bernadino',
                  'santa_fe',
                  'dallas', 
                  'champaign',
                  'mansfield',
                  'syracuse']
```

```python
# get the list of files
wind_files_location = {}
solar_files_location = {}

for loc in solar_locations:
    files = glob.glob("./"+loc+"/*.csv")
    solar_files_location[loc] = files
    
for i in solar_files_location.values():
    i.sort()
```

```python
solar_files_location
```

```python
for loc in solar_files_location:
#     print(loc)
    print(f"Description of data for {loc.capitalize()}")
    for i, file in enumerate(solar_files_location[loc]):
#         print(i, file)
        if i == 0:
            df = pd.read_csv(file, 
                         skiprows=2,)
            df.index = pd.to_datetime(df[['Year', 'Month', 'Day', 'Hour', 'Minute']])
            df.drop(['Year', 'Month', 'Day', 'Hour', 'Minute'], axis=1, inplace=True)
#             print(df.head())
        else:
            tmp_df = pd.read_csv(file, 
                         skiprows=2,)
            tmp_df.index = pd.to_datetime(tmp_df[['Year', 'Month', 'Day', 'Hour', 'Minute']])
            tmp_df.drop(['Year', 'Month', 'Day', 'Hour', 'Minute'], axis=1, inplace=True)
#             print(tmp_df.head())
            df = pd.concat([df, tmp_df], axis=0)
    df.to_csv('./'+loc+'_complete.csv')
    print(df.describe())
```
